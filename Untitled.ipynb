```r
library(dplyr)
library(ggplot2)
library(tidyverse)
```

```r
df <- read.csv(
    "liss_data.tsv", 
    sep="\t", 
    col.names=c("prob", "order", "algo_time", "lp_time", "integrality_gap")
)
```

```r
df_tidy <- df %>% tidyr::gather(key="Algo", value="Time", -prob, -order, -integrality_gap)
df_tidy$prob = as.ordered(df_tidy$prob)
df_tidy$order = as.ordered(df_tidy$order)
df_tidy$Algo = as.factor(df_tidy$Algo)
head(df_tidy)
dim(df)
dim(df_tidy)
```

```r
df_tidy %>% 
    filter(Algo == "algo_time") %>% 
    ggplot() +
    aes(x=prob, y=Time, group=order, color=order) +
    stat_summary(fun.y = mean, geom="point") +
    stat_summary(fun.y = mean, geom="line") +
    xlab("Erdos Renyi Probability") +
    ylab("Average Time Taken (sec)") +
    ggtitle("Largest Independent Set: Tsukiyama et al.")
```

```r
df_tidy %>% 
    filter(Algo == "lp_time") %>% 
    ggplot() +
    aes(x=prob, y=Time, group=order, color=order) +
    stat_summary(fun.y = mean, geom="point") +
    stat_summary(fun.y = mean, geom="line") +
    xlab("Erdos Renyi Probability (p)") +
    ylab("Average Time Taken (sec)") +
    ggtitle("Largest Independent Set: Gurobi LP Solver")
```

```r
df_tidy %>% 
    ggplot() +
    aes(x=prob, y=Time, group=Algo, color=Algo) +
    stat_summary(fun.y = mean, geom="point") +
    stat_summary(fun.y = mean, geom="line") +
    xlab("Erdos Renyi Probability (p)") +
    ylab("Average Time Taken (sec)") +
    ggtitle("Largest Independent Set: Tsukiyama et al v/s LP Solver (Gurobi)")
```

```r
df_tidy %>% 
    ggplot() +
    aes(x=prob, y=integrality_gap, group=Algo, color=Algo) +
    stat_summary(fun.y = mean, geom="point") +
    stat_summary(fun.y = mean, geom="line") +
    xlab("Erdos Renyi Probability (p)") +
    ylab("Integrality Gap") +
    ggtitle("Largest Independent Set: Integrality Gap")
```

```r
wvc_df <- read.csv("weighted_vertex_cover_new.tsv", sep="\t", col.names = c("prob", "order", "lp_time", "ilp_time", "integrality_gap"))
wvc_df$prob = as.ordered(wvc_df$prob)
wvc_df$order = as.ordered(wvc_df$order)
head(wvc_df)
```

```r
vc_df <- read.csv("sage_vertex_cover.tsv", sep=" ", col.names = c("prob", "order", "algo_time", "lp_time"))
vc_df$prob = as.ordered(vc_df$prob)
vc_df$order = as.ordered(vc_df$order)

vc_df_tidy <- vc_df %>% tidyr::gather(key="Method", value="time_taken", -prob, -order)
head(vc_df)
head(vc_df_tidy)
vc_df_tidy$Method = as.factor(vc_df_tidy$Method)
```

```r
vc_df_tidy %>% 
    ggplot() +
    aes(x=prob, y=time_taken, group=Method, color=Method) +
    stat_summary(fun.y = mean, geom="point") +
    stat_summary(fun.y = mean, geom="line") +
    xlab("Erdos Renyi Probability (p)") +
    ylab("Time Taken") +
    ggtitle("Vertex Cover for Random Graphs")
```

```r
vc_df_tidy %>% 
    filter(Method == 'algo_time') %>%
    ggplot() +
    aes(x=prob, y=time_taken, group=order, color=order) +
    stat_summary(fun.y = mean, geom="point") +
    stat_summary(fun.y = mean, geom="line") +
    xlab("Erdos Renyi Probability (p)") +
    ylab("Time Taken") +
    ggtitle("Vertex Cover for Random Graphs: Cliquer Algorithm")
```

```r
vc_df_tidy %>% 
    filter(Method == 'lp_time') %>%
    ggplot() +
    aes(x=prob, y=time_taken, group=order, color=order) +
    stat_summary(fun.y = mean, geom="point") +
    stat_summary(fun.y = mean, geom="line") +
    xlab("Erdos Renyi Probability (p)") +
    ylab("Time Taken") +
    ggtitle("Vertex Cover for Random Graphs: LP Solver")
```
